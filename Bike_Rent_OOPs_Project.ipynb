Define a function `rentBike`, it should have following conditions:

- if `number_of_bikes` is less than or equal to zero, it should display the message "Number of bikes should be positive!".

- if `number_of_bikes` is greater than the `stock`, it should display the `number_of_bikes` available to rent.

- Otherwise display the total `number_of_bikes` rented by the customer. Do not forget to update the value of `stock` i.e. subtract the `number_of_bikes` rented from stock and display the updated `stock`.

```python
def rentalBike(number_of_bikes,stock):
    if number_of_bikes <= 0:
        print("Number of bikes should be positive:")
    elif number_of_bikes > stock:
        print(f"The number of bikes avaliable to rent is {stock}")
    else:
        print(f"Total number of bill rented by the customer: {number_of_bike}")
        stock -= number_of_bikes
        print(f"Update stock:{stock}")
    return stock
```

Define a function `returnBike` to return the total bill of customer.
- Check if `rentalBasis` is ' daily ' or ' weekly ' and accordingly calculate the bill. Rent on ' daily ' basis is Rs. 100 per bike and for ' weekly ' basis it is Rs. 500 per bike.

- If `number_of_bikes` rented by customer is between 3 to 5 (both values inclusive), 30% discount is offered on total bill. So it should display the message "You are eligible for Family rental promotion of 30% discount".

```python
def returnBike(number_of_bikes,rental_basis):
    daily_rate = 100
    weekly_rate = 500
    bill = 0

    if rental_basis == "daily":
        bill = number_of_bikes * daily_rate
    elif rental_basis == "weekly":
        bill = number_of_bikes + weekly_rate

    if 3 <= number_of_bikes <= 5:
        discount = 0.30
        bill = bill *(1 - discount)
        print("You are eligible for family rental promotion of 30 % discount")

    return bill
```

## Step 3

Write complete `Customer` class using above defined functions `returnBike` and `rentBike` inside it as methods.

```python
class Customer(BikeRental):
    bill = 0 # Class attribute to track total revenue

    def __init__(self, rental_time, rental_basis, number_of_bikes):
        self.rental_time = rental_time
        self.rental_basis = rental_basis
        self.number_of_bikes = number_of_bikes

    def rentBike(self):
        if self.number_of_bikes <= 0:
            print("Number of bikes should be positive!")
        elif self.number_of_bikes > BikeRental.stock:
            print(f"The number of bikes available to rent is {BikeRental.stock}")
        else:
            print(f"Total number of bikes rented by the customer: {self.number_of_bikes}")
            BikeRental.stock -= self.number_of_bikes
            print(f"Updated stock: {BikeRental.stock}")

    def returnBike(self):
        global daily_rate, weekly_rate # Access global rates

        calculated_bill = 0
        if self.rental_basis == "daily":
            calculated_bill = self.number_of_bikes * daily_rate * self.rental_time
        elif self.rental_basis == "weekly":
            calculated_bill = self.number_of_bikes * weekly_rate * self.rental_time # Corrected from + to *

        if 3 <= self.number_of_bikes <= 5:
            discount = 0.30
            calculated_bill = calculated_bill * (1 - discount)
            print("You are eligible for Family rental promotion of 30% discount")

        Customer.bill += calculated_bill # Update the class attribute for total revenue
        print(f"Your total bill for this rental: Rs.{calculated_bill:.2f}")

        # Return bikes to stock
        BikeRental.stock += self.number_of_bikes
        print(f"Bikes returned. Updated stock: {BikeRental.stock}")

        return calculated_bill
```
